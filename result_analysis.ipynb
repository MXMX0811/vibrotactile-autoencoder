```python
import os
import librosa
import librosa.display
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
R_img_dir = 'Reference_Waves'
data, fs = librosa.load(R_img_dir + '/G2_2.wav', sr=44100)
data = data[:int(len(data)/10)]
plt.plot(data)
```

```python
DMatrix = librosa.stft(data, n_fft=2048, hop_length=int(2048 * 0.1), window='hann')
dbMatrix = librosa.amplitude_to_db(np.abs(DMatrix), ref=np.max)
fig, ax = plt.subplots()
img = librosa.display.specshow(dbMatrix, sr=fs, n_fft=2048, hop_length=int(2048 * 0.1), y_axis='linear', x_axis='time', ax=ax, cmap='jet')
ax.set_title('Spectrogram')
plt.ylim(0, 1000)
fig.colorbar(img, ax=ax, format="%+2.0f dB")
```

```python
G_img_dir = 'Generated_Waves'
data, fs = librosa.load(G_img_dir + '/0118/G2_2_20240118183305.wav', sr=44100)
data = data[:int(len(data)/10)]
plt.plot(data)
```

```python
DMatrix = librosa.stft(data, n_fft=2048, hop_length=int(2048 * 0.1), window='hann')
dbMatrix = librosa.amplitude_to_db(np.abs(DMatrix), ref=np.max)
fig, ax = plt.subplots()
img = librosa.display.specshow(dbMatrix, sr=fs, n_fft=2048, hop_length=int(2048 * 0.1), y_axis='linear', x_axis='time', ax=ax, cmap='jet')
ax.set_title('Spectrogram')
plt.ylim(0, 1000)
fig.colorbar(img, ax=ax, format="%+2.0f dB")
```

```python
conf_matrix_g = np.zeros([5, 5], dtype=np.int64)
results_path = 'Evaluation_Results'
dir_list = os.listdir(results_path)
for dir in dir_list:
    if os.path.basename(dir)[0:8] == 'cm_cross':
        file = os.path.join(results_path, dir)
        conf_matrix_g += np.array(pd.read_csv(file, index_col=0))
```

```python
conf_matrix_r = np.zeros([5, 5], dtype=np.int64)
results_path = 'Evaluation_Results'
dir_list = os.listdir(results_path)
for dir in dir_list:
    if os.path.basename(dir)[0:7] == 'cm_real':
        file = os.path.join(results_path, dir)
        conf_matrix_r += np.array(pd.read_csv(file, index_col=0))
```

```python
cm = ConfusionMatrixDisplay(confusion_matrix=conf_matrix_g,
                            display_labels=['G2', 'G3', 'G4', 'G6', 'G8'])

print(np.diagonal(conf_matrix_g).sum() / conf_matrix_g.sum())

fig, ax = plt.subplots(figsize=(6, 6))
plt.title('Confusion Matrix of Generated Vibrations')

cm.plot(
    include_values=True,
    cmap="Blues",
    colorbar=False,
    ax=ax,
    xticks_rotation="horizontal",
    values_format="d"
)
```

```python
cm = ConfusionMatrixDisplay(confusion_matrix=conf_matrix_r,
                            display_labels=['G2', 'G3', 'G4', 'G6', 'G8'])

print(np.diagonal(conf_matrix_r).sum() / conf_matrix_r.sum())

fig, ax = plt.subplots(figsize=(6, 6))
plt.title('Confusion Matrix of Real Vibrations')

cm.plot(
    include_values=True,
    cmap="Blues",
    colorbar=False,
    ax=ax,
    xticks_rotation="horizontal",
    values_format="d"
)
```
